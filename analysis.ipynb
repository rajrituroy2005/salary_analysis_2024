````python
def format_salary(x):
    if x >= 1_000_000:
        return f"{x/1_000_000:.2f}M".rstrip('0').rstrip('.')
    elif x >= 1_000:
        return f"{x/1_000:.2f}K".rstrip('0').rstrip('.')
    else:
        return str(x)

df['salary'] = df['salary'].apply(format_salary)
````

````python
def convert_to_number(x):
    if isinstance(x, str):
        x = x.strip().upper()

        if x.endswith('K'):
            return int(x[:-1]) * 1000
        elif x.endswith('M'):
            return int(x[:-1]) * 1_000_000
        else:
            return int(x)  # if no K/M, keep as number
    return x

df['salary'] = df['salary'].apply(convert_to_number)

````

````python
df['salary'] = df['salary'].astype(int)
````

# 2024 Salary Analysis Report

## 1. Overview

This report explores a 2024 salary dataset containing **16,534 records** and the following key columns:

- `work_year`  Year of the salary data (2024)
- `experience_level`  Experience band (Entry-level, Mid-level, Senior-level, Executive-level)
- `employment_type`  Contract type (Full-time, Part-time, Contract, Freelance)
- `job_title`  Role title (e.g., Data Engineer, AI Engineer)
- `salary`  Annual salary (numeric)
- `salary_currency`  Original currency of reported salary
- `employee_residence`  Country where the employee lives
- `remote_ratio`  Percentage of remote work (0 = onsite, 50 = hybrid, 100 = fully remote)
- `company_location`  Country where the company is based
- `company_size`  Size of the company (Small, Medium, Large)

The goal of this analysis is to answer core questions about **compensation patterns** across roles, experience, locations, employment types, and company sizes.
